```python
#make_classification (2)

from sklearn.datasets import make_classification 
import pandas as pd

x, y = make_classification(n_samples=100, 
                           n_features=6, n_informative=3, 
                           n_redundant=2, n_repeated=0, 
                           n_clusters_per_class=2,
                           n_classes=3, 
                           random_state=0)

#ถ้าต้องการแปลงเป็น DataFrame
df = pd.DataFrame(x, columns=['x1', 'x2', 'x3', 'x4', 'x5', 'x6'])
df['y'] = y

with pd.option_context('display.max_rows', 6): display(df)
```

```python
#make_regression

from sklearn.datasets import make_regression 
import pandas as pd

x, y = make_regression(n_samples=100, n_features=5, random_state=0)

#ถ้าต้องการแปลงเป็น DataFrame
df = pd.DataFrame(x, columns=['x1', 'x2', 'x3', 'x4', 'x5'])
df['y'] = y

with pd.option_context('display.max_rows', 10): display(df)
```

```python
#make_blobs

from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x, y = make_blobs(n_samples=100, centers=3, random_state=1)

df = pd.DataFrame(x, columns=['x1', 'x2'])
df['y'] = y

with pd.option_context('display.max_rows', 6): display(df)

colors = np.array(['#999', '000', '#ddd'])
plt.scatter(df['x1'], df['x2'], color=colors[df['y']])
plt.show()
```

```python
#make_regression

from sklearn.datasets import make_circles
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x, y = make_circles(n_samples=100, noise=0.08, random_state=0)

df = pd.DataFrame(x, columns=['x1', 'x2'])
df['y'] = y

with pd.option_context('display.max_rows', 6): display(df)

colors = np.array(['#999', '000'])
plt.scatter(df['x1'], df['x2'], color=colors[df['y']])
plt.show()
```

```python
#label encoder 

from sklearn.preprocessing import LabelEncoder

data = ['Apple', 'Mango', 'Grape', 'Durian', 'Mango', 'Grape']
encoder = LabelEncoder()

encoder.fit(data)

enc_data1 = encoder.transform(data)
print(enc_data1)
#[0 3 2 1 3 2]

#ถ้าจะแปลงข้อมูลครั้งต่อไป (หากเป็นชุดเดียวกัน) ให้เรียกเมธอด transform() ได้เลย
enc_data2 = encoder.transform(['Mango', 'Durian', 'Durian'])
print(enc_data2)
#[3 1 1]

enc_data3 = encoder.transform(['Grape', 'Mango', 'Mango', 'Durian'])
print(enc_data3)
#[2 3 3 1]

enc_data4 = encoder.transform(['mango', 'Mangosteen', 'Melon'])  #Error
#เกิดข้อผิดพลาด เพราะ
#mango เขียนด้วยตัวพิมพ์ที่ไม่ตรงกับข้อมูลที่กำหนดให้แก่ fit()
#Mangosteen และ Melon ไม่มีอยู่ในข้อมูลที่กำหนดให้แก่ fit()
```

```python
#label encoder 

from sklearn.preprocessing import LabelEncoder

data = ['Apple', 'Mango', 'Grape', 'Durian', 'Mango', 'Grape']
encoder = LabelEncoder()

enc_data1 = encoder.fit_transform(data)
print(enc_data1)

enc_data2 = encoder.transform(['Mango', 'Durian', 'Durian'])
print(enc_data2)

enc_data2 = encoder.transform(['Apple', 'Grape', 'Durian', 'Grape'])
print(enc_data2)
```

```python
#standard scaler

from sklearn.preprocessing import StandardScaler

data = [[2, 2000],
        [3.5, 5500],
        [1.75, 2150],
        [3.14, 4130],
        [2.25, 2564]]

scaler = StandardScaler()
scaler.fit(data)
data2 = scaler.transform(data)
print(data2)
print()

data_test = [[1.23, 3210]]
print(scaler.transform(data_test))
```

```python
#standard scaler

from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

scaler = StandardScaler()

df = pd.read_csv(r'data\test-standard-scaler.csv')
display(df)

#ถ้าใช้วิธีเรียก fit() และตามด้วย transform()
'''
scaler.fit(df[['volume', 'weight']])
df[['volume', 'weight']] = scaler.transform(df[['volume', 'weight']])
display(df)
'''

#ถ้าใช้วิธีเรียก fit_transform() ในขั้นตอนเดียว
df[['volume', 'weight']] = scaler.fit_transform(df[['volume', 'weight']])
display(df)

#ถ้าต้องการปรับสเกลข้อมูลเพิ่มเติม 
#ต้องกำหนดในแบบลิสต์หรืออาร์เรย์ 2 มิติ
#และต้องมีจำนวนคอลัมน์เท่ากับที่กำหนดให้แก่ fit() หรือ fit_transform()
data = [[1.9, 1000]]
print(scaler.transform(data))

print(scaler.transform([[2.3, 1234]]))
```

```python
#train_test_split (1)

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

x, y = make_classification(n_samples=10, n_features=4, 
                           n_classes=2, random_state=0)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

print(x_train)
print(y_train)
print()
print(x_test)
print(y_test)
```
