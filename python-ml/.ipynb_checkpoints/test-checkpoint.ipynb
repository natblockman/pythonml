```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(1)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)

plt.scatter(X_xor[y_xor == 1, 0],X_xor[y_xor == 1, 1], c='w', marker='o', edgecolors='r', label='1')
plt.scatter(X_xor[y_xor == -1, 0],X_xor[y_xor == -1, 1],c='r',marker='o',label='-1')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.legend(loc='best')
plt.show()

data = {'x1':X_xor[:,0], 'x2':X_xor[:,1], 'y':y_xor}

df = pd.DataFrame(data)
idx = df[(df['x1'] < 0) & (df['x2'] < 0) & (df['y'] == -1)].index
df.drop(idx, inplace=True)

df.to_csv(r'data\svm_poly2.csv', index=None)
```

```python
import numpy as np

inputs = [101, 108]
weights_h1 = [0.15, 1]
weights_h2 = [1, -1.25]
weights_h3 = [0.2, -0.2]

sum_h1 = np.dot(inputs, weights_h1) + 1   
sum_h2 = np.dot(inputs, weights_h2) + 2.5 
sum_h3 = np.dot(inputs, weights_h3) + 1.5 

out_h1 = max(0, sum_h1)
out_h2 = max(0, sum_h2)
out_h3 = max(0, sum_h3)

sum_h4 = (out_h1 * 0.15) + (out_h2 * 1.5) + (out_h3 * 2.5) + 1
sum_h5 = (out_h1 * 0.1) + (out_h2 * 2) + (out_h3 * 1) + 2

out_h4 = max(0, sum_h4)
out_h5 = max(0, sum_h5)
 
sum_out = (out_h4 * 1) + (out_h5 * 1) + 0
output = 1 / (1 + np.exp(-sum_out))

print('{:.2f}'.format(output))
```

```python
#example 10-14

from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_excel(r'data\car_miles_per_gallon.xlsx')
df.dropna(inplace=True)
#with pd.option_context('display.max_rows', 10): display(df)

#คอลัมน์ที่จะใช้ Train Model: 
#Cylinders, Displacement, Horsepower, Weight, Acceleration
x = df.iloc[:, 3:8]  
y = df['MPG']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

model = MLPRegressor(hidden_layer_sizes=(4), #Hidden1 มี 4 นิวรอน
                      activation='relu',
                      solver='lbfgs',
                      max_iter=10000,
                      random_state=0)

model.fit(x_train, y_train)

#ข้อมูลที่จะทำนายผล ก็ต้องแปลงสเกลเช่นกัน
x_predict = [[6, 400, 150, 4000, 9]]
x_predict = scaler.transform(x_predict)

predicted = model.predict(x_predict)

print('Predicted:', '{:.2f}'.format(predicted[0]))
print('Score:', '{:.2f}'.format(model.score(x_test, y_test)), '\n')

print('weight:')
print(*model.coefs_, sep='\n')

print()
print('bias:')
print(*model.intercepts_, sep='\n')
```

```python
from numpy import mean 
from numpy import std 
from sklearn.datasets import make_classification 
from sklearn.model_selection import cross_val_score 
from sklearn.model_selection import RepeatedStratifiedKFold 
from sklearn.linear_model import LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.svm import SVC 
from sklearn.naive_bayes import GaussianNB 
from sklearn.ensemble import StackingClassifier 
from matplotlib import pyplot 
from sklearn.datasets import load_wine,load_iris 
from matplotlib.pyplot import figure 

figure(num=2, figsize=(16, 12), dpi=80, facecolor='w', edgecolor='k')       
# get a stacking ensemble of models 
def get_stacking():  
    # define the base models   
    level0 = list()   
    level0.append(('lr', LogisticRegression()))   
    level0.append(('knn', KNeighborsClassifier()))  
    level0.append(('cart', DecisionTreeClassifier()))   
    level0.append(('svm', SVC()))   
    level0.append(('bayes', GaussianNB()))   
    # define meta learner model   
    level1 = LogisticRegression()   
    # define the stacking ensemble   
    model = StackingClassifier(estimators=level0, final_estimator=level1, cv=5)   
    return model   

# get a list of models to evaluate 
def get_models():   
    models = dict()   
    models['LogisticRegression'] = LogisticRegression()   
    models['KNeighborsClassifier'] = KNeighborsClassifier()   
    models['Decision tree'] = DecisionTreeClassifier()   
    models['svm'] = SVC()   
    models['GaussianNB'] = GaussianNB()   
    models['stacking'] = get_stacking()   
    return models   

# evaluate a give model using cross-validation 
def evaluate_model(model):   
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)   
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')   
    scores1 = cross_val_score(model, X1, y1, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')   
    return scores,scores1   

# define dataset 
X,y = load_wine().data,load_wine().target 
X1,y1= load_iris().data,load_iris().target 
# get the models to evaluate 
models = get_models() 
# evaluate the models and store results 
results, names, results1 = list(), list(),list() 
for name, model in models.items():
    scores,scores1= evaluate_model(model)   
    results.append(scores)   
    results1.append(scores1)   
    names.append(name)   
    print('>%s -> %.3f (%.3f)---Wine dataset' % (name, mean(scores), std(scores)))   
    print('>%s -> %.3f (%.3f)---Iris dataset' % (name, mean(scores1), std(scores1))) 
    # plot model performance for comparison 
    pyplot.rcParams["figure.figsize"] = (15,6) 
    pyplot.boxplot(results, labels=[s+"-wine" for s in names], showmeans=True) 
    pyplot.show() pyplot.boxplot(results1, labels=[s+"-iris" for s in names], showmeans=True) 
    pyplot.show()
    
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

x, y = make_classification(n_samples=700, n_features=4, n_classes=2, random_state=1)

df = pd.DataFrame(x, columns=['x1', 'x2', 'x3', 'x4'])
df['y'] = y

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)

#KNN
knn = KNeighborsClassifier()
knn_params = {'n_neighbors': range(1, 50, 2)}
knn_rscv = RandomizedSearchCV(estimator=knn, param_distributions=knn_params)
knn_rscv.fit(x, y)

print('KNN Score:', knn_rscv.score(x_test, y_test))
knn_best = knn_rscv.best_estimator_
#---------------------------------------

#Random Forest
forest = RandomForestClassifier()
forest_params = {
    'n_estimators': range(10, 100, 50),
    'criterion': ['entropy', 'gini']
}
forest_rscv = RandomizedSearchCV(estimator=forest, param_distributions=forest_params, n_iter=4)
forest_rscv.fit(x, y)

print('Random Forest Score:', forest_rscv.score(x_test, y_test))
forest_best = forest_rscv.best_estimator_
#---------------------------------------

#SVM
svm = SVC()
svm_params = [
    {'kernel': ['linear'], 'C': [0.1, 1, 5, 10, 50, 100]},
    {'kernel': ['rbf'], 'C': [0.1, 1, 5, 10, 50, 100], 'gamma': range(1, 10)},
    {'kernel': ['poly'], 'C': [0.1, 1, 5, 10, 50, 100], 'degree': range(1, 10)},
]

svm_rscv = RandomizedSearchCV(estimator=svm, param_distributions=svm_params)
svm_rscv.fit(x, y)

print('SVM Score:', svm_rscv.score(x_test, y_test))
svm_best = svm_rscv.best_estimator_
#---------------------------------------

df = pd.DataFrame(x, columns=['x1', 'x2', 'x3', 'x4'])
df['y'] = y

#with pd.option_context('display.max_rows', 10): display(df)

estimators = [
    ('knn', knn_best),
    ('forest', forest_best),
    ('svm', svm_best),
]

voting = VotingClassifier(estimators, voting='hard')  #voting='hard' เป็นค่าดีฟอลต์อยู๋แล้ว

voting.fit(x_train, y_train)
print('Voting Score:', voting.score(x_test, y_test))


stack = StackingClassifier(estimators, 
                           final_estimator=LogisticRegression())

stack.fit(x_train, y_train)
print('Stacking Score:', stack.score(x_test, y_test))
```

```python
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder 
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

df = pd.read_excel(r'data\animals.xlsx')
#display(df)

#ข้อมูลมีทั้งหมด 7 คอลัมน์
#'Name', 'Blood_Temperature', 'Give_Birth', 'Can_Fly', 
#'Live_In_Water', 'Have_Legs', 'Species'

df_x = df.iloc[:, 1:6]  #เลือกทุกแถว, คอลัมน์ที่ 1 - 5 (ไม่เลือกคอลัมน์ลำดับ 0 และ 6)
df_y = df['Species']

enc_dict = defaultdict(LabelEncoder)
df_x = df_x.apply(lambda v: enc_dict[v.name].fit_transform(v))

model = DecisionTreeClassifier(criterion='entropy')
model.fit(df_x, df_y)

#Blood_Tem, Give_Birth, Can_Fly, Live_In_Water, Have_Legs
df_predict = pd.DataFrame([['Warm', 'No', 'Yes', 'Sometimes', 'No']], 
                          columns=df_x.columns)

df_predict = df_predict.apply(lambda v: enc_dict[v.name].transform(v))

predicted = model.predict(df_predict)
print('Predicted:', predicted[0])

score = model.score(df_x, df_y)
print('Accuracy:', '{:.2f}'.format(score))
```
