```python
import numpy as np
import pandas as pd

data = { 'One': np.random.randint(100, 1000, 10),
         'Two': np.random.uniform(1, 10, 10),
         'Three': np.random.rand(10),
         'Four': np.random.randint(1, 100, 10),
         'Five': np.random.uniform(-10, 10, 10) }

df = pd.DataFrame(data)

display(df[['One', 'Three', 'Five']],
        df[:][['Two', 'Three', 'Four']],
        df.iloc[2:5, 3:5],
        df.iloc[::, [0, 2, 4]])
```

```python
import numpy as np
import pandas as pd

data = { 'One': np.random.randint(100, 1000, 10),
         'Two': np.random.uniform(1, 10, 10),
         'Three': np.random.rand(10),
         'Four': np.random.randint(1, 100, 10),
         'Five': np.random.uniform(-10, 10, 10) }

df = pd.DataFrame(data)

display(
    df.sum(),
    df.sum()['One'],
    df.max(),
    df.max()['One'],
    df.describe()
)

df.info()
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')
y = (15, 12, 19, 13, 16, 17, 12)
x = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'] 
plt.title('Weekly Sale', color='blue', alpha=0.8)
plt.ylabel('Amount', color='m', fontsize=14)
plt.xlabel('Weekday', color='g', fontsize=14)
plt.yticks(color='r')
plt.xticks(rotation='vertical', color='r', fontsize=13)
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')

y = [10, 15, 15, 20, 10]
x = [1, 2, 3, 4, 5]
plt.plot(x, y, ls='-', c='r', label='Product #1')

y = [15, 25, 10, 15, 20]
plt.plot(x, y, ls='--', c='g', label=' Product #2')

y = [10, 20, 15, 10, 15]
plt.plot(x, y, ls=':', c='b', label=' Product #3')
  
plt.legend(loc='best')
plt.show()
```
