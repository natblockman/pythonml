```python
#example 11-2

from sklearn.cluster import KMeans

data = [[2, 2], [4, 6], [3, 5], [7, 2], [6, 6], 
        [3, 9], [8, 8], [8, 4], [4, 3], [5, 5]]

model = KMeans(n_clusters=2, n_init=200, random_state=0)
model.fit(data)

print(model.labels_)
print(model.cluster_centers_)
print()

pd = model.predict([[7, 7]])     #ทำนายผลของตำแหน่ง (7, 7)
print(pd[0])
```

```python
#example 11-4

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

d = [[2, 2], [4, 6], [3, 5], [7, 2], [6, 6], 
     [3, 9], [8, 8], [8, 4], [4, 3], [5, 5]]

data = np.array(d)  #ถ้าเปลี่ยนจากลิสต์เป็น Numpy Array จะจัดการได้ง่ายกว่า

k = 2
model = KMeans(n_clusters=k, n_init=200, random_state=0)
model.fit(data)

predict = model.predict(data)  #ทำนายผลของข้อมูลทั้งหมด

c = ['red', 'white']

colors = []
edge_colors = []

for i in predict:
    colors.append(c[i])
    edge_colors.append('black' if c[i] == 'white' else c[i])
    
#plot ข้อมูลทั้งหมด ตามสีที่ได้กำหนดเอาไว้
plt.scatter(data[:, 0], data[:, 1], c=colors, edgecolor=edge_colors, s=80)

#plot จุด centroid ให้เป็นรูปดาว
centroids = model.cluster_centers_

#ข้อมูลที่ได้จาก cluster_centers_ เป็นอาร์เรย์ในรูปแบบ
#[[x1, y1], [x2, y2], ...]
#เราสามารถใช้วิธีการของ numpy เพื่ออ้างถึงข้อมูลแต่ละตำแหน่ง
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=150)

plt.show()
```

```python
#example 11-5

from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'data\k-means.csv')
#with pd.option_context('display.max_rows', 10): display(df)

model = KMeans(n_clusters=3, random_state=0)
model.fit(df)

predict = model.predict(df) 

c = ['red', 'white', '#ddd']
colors = []
edge_colors = []

for i in predict:
    colors.append(c[i])
    edge_colors.append('black' if c[i] == 'white' else c[i])

centroids = model.cluster_centers_

plt.scatter(df['x'], df['y'], c=colors, edgecolor=edge_colors, s=80)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=150)
plt.show()
```

```python
#example 11-6

import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_csv(r'data\bmx.csv')

#เฉพาะคอลัมน์ bmxleg และ bmxwaist
#โดยในบางเซลล์ อาจเป็นค่า nan ซึ่งเราควรลบแถวที่มีเซลล์ดังกล่าวออกไป
df = df[['bmxleg', 'bmxwaist']].dropna()

#with pd.option_context('display.max_rows', 10): display(x)

model = KMeans(n_clusters=4)
model.fit(df)

centroids = model.cluster_centers_
print('Centroids:\n', centroids)

#นำข้อมูลแต่ละตำแหน่งไปทำนายว่าอยู่กลุ่มใด
#ซึ่งจะได้ผลลัพธ์เป็นอาร์เรย์ของลำดับกลุ่มของข้อมูลแต่ละตำแหน่ง
clusters = model.predict(df)

#กำหนดสีของแต่ละกลุ่ม (ตามค่า K หรือ n_clusters)
cluster_colors = ['red', 'green', 'blue', 'lightblue']

#เก็บสีของข้อมูลในแต่ละตำแหน่งตามกลุ่มของมัน
data_colors = []
for i in clusters:
    data_colors.append(cluster_colors[i])

plt.figure(figsize=(10, 6))

#วาดตำแหน่งข้อมูล โดยให้แกน X เป็นค่า bxmleg และแกน Y เป็น bmxwaist
#สีของแต่ละตำแหน่งตามค่าที่เก็บไว้ในลิสต์
plt.scatter(df['bmxleg'], df['bmxwaist'], c=data_colors)

#วาดตำแหน่ง Centroid
plt.scatter(centroids[:, 0], centroids[:, 1], 
            marker='*', color='black', s=150)   #s = marker size

plt.show()
```

```python
#example 11-7

from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics

data = [[2, 2], [4, 6], [3, 5], [7, 2], [6, 6], 
        [3, 9], [8, 8], [8, 4], [4, 3], [5, 5]]

silhouette_avgs = []
min_k = 2
max_k = len(data)

for k in range(min_k, max_k):
    model = KMeans(n_clusters=k, n_init=200, random_state=0)
    model.fit(data)
    s_score = metrics.silhouette_score(data, model.labels_)
    print("Silhouette Coeff for k =", k, "is", s_score)
    silhouette_avgs.append(s_score)

#optimal k คือกรณีที่ให้ค่า silhouette coeff สูงสุด
optimal_k = silhouette_avgs.index(max(silhouette_avgs)) + min_k
print("Optimal K is ", optimal_k)

#นำมา plot กราฟเพื่อให้เห็นการเปรียบเทียบที่ชัดเจน
plt.plot(range(min_k, len(data)), silhouette_avgs)
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
#example 11-8

from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics

df = pd.read_csv(r'data\k-means.csv')
#with pd.option_context('display.max_rows', 10): display(df)

silhouette_avgs = []
min_k = 2
max_k = 13  #ถ้าจะให้ครบจำนวนรายการข้อมูลต้องใช้ len(df)  

for k in range(min_k, max_k):  
    model = KMeans(n_clusters=k,random_state=0).fit(df)
    s_score = metrics.silhouette_score(df, model.labels_)
    print('Silhouette Coeff for k =', k, 'is', s_score)
    silhouette_avgs.append(s_score)

optimal_k = silhouette_avgs.index(max(silhouette_avgs)) + min_k
print('Optimal K is ', optimal_k)

#plt.subplots(figsize=(7, 5))
plt.plot(range(min_k, max_k), silhouette_avgs)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.xticks(range(min_k, max_k))
plt.show()
```
